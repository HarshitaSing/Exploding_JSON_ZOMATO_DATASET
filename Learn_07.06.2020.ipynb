```
#Load the JSON data
restaurant_df = spark.read.json('/FileStore/tables/file3.json')
restaurant_df.count()
```

```
#Print the schema
restaurant_df.printSchema()
```

```
#Exploding Restaurants attribute from primary dataset(restaurant_df)
from pyspark.sql.functions import explode

explodedDF = restaurant_df.select(explode(restaurant_df.restaurants))
display(explodedDF.limit(5))
```

```
#Selecting required attributes from the schema

df_child = restaurant_df.select(
        "restaurants.restaurant.R",
        "restaurants.restaurant.apikey",
  "restaurants.restaurant.name",
  "restaurants.restaurant.cuisines",
  "restaurants.restaurant.average_cost_for_two",
  "restaurants.restaurant.book_url",
  "restaurants.restaurant.location",
  "restaurants.restaurant.offers", 
  "restaurants.restaurant.price_range",
  "restaurants.restaurant.user_rating",
  "restaurants.restaurant.has_online_delivery"
)
df_child.printSchema()
```

```
#Count of the dataset

df_child.count()
```

```
#Exploding arrayType columns together

from pyspark.sql.functions import arrays_zip, col,explode

exploded_attributes = df_child.withColumn("tmp",arrays_zip("offers","user_rating","R")).withColumn("tmp",explode("tmp")).select( col("tmp.R"),"cuisines",col("tmp.offers"), col("tmp.user_rating"),"name","location","price_range", "has_online_delivery", "book_url")
    
display(exploded_attributes.limit(10))
```

```
#Count of exploded dataset

exploded_attributes.count()
```

```
#Dropping the null columns

exploded_attributes = exploded_attributes.drop("book_url","offers","has_online_delivery")
exploded_attributes.printSchema()
```

```
#Duplicates Check

exploded_attributes.dropDuplicates()
```

```
#Count of Null values in the dataset

from pyspark.sql.functions import when, count, col
null_count = exploded_attributes.select([count(when(col(c).isNull(), c)).alias(c) for c in 
           exploded_attributes.columns])
display(null_count)
```

```
#Displaying dataset
exploded_attributes.show(2)
```

```
exploded_attributes.printSchema()
```

```
#converting the dataset to dataframe

exploded_df = exploded_attributes.toDF("R", "cusisines","user_rating","name","location","price_range").select("*")
type(exploded_df)
```

```
exploded_df.printSchema()
```

```
#Filtering data on a single ID (unique)
from pyspark.sql.functions import col

filtered_df_16608059 = exploded_df.filter(col("R.res_id") == 16608059)
```

```
filtered_df_16608059.count()
```

```
#Display filtered data

display(filtered_df_16608059)
```

```
#Dropping Duplicates.

filtered_df_16608059 = filtered_df_16608059.dropDuplicates()
```

```
#filtered_df_16608059.count()
display(filtered_df_16608059)
```

```
#Exploding the arrayType columns
from pyspark.sql.functions import arrays_zip, col,explode

exploded_16608059_df = filtered_df_16608059\
.withColumn("temp",arrays_zip("cusisines","name","location","price_range"))\
.withColumn("temp",explode("temp"))\
.select("R",col("temp.cusisines"),"user_rating",col("temp.name"),col("temp.location"),col("temp.price_range"))

display(exploded_16608059_df.limit(5))
```

```
exploded_16608059_df_new = exploded_16608059_df.dropDuplicates()
exploded_16608059_df_new.count()
```
